```python
#Bai 1
n = list(map(int,input().split()))
x = []
for i in n:
  if i not in x:
    x.append(i)
print(x)
```

```python
#Bai 3
n = list(map(int,input().split()))
k = int(input())
lst = []
for i in range(k,len(n)):
  lst.append(n[i])
for i in range(0,k):
  lst.append(n[i])
print(tuple(lst))

```

```python
#Bai 4
kv = list(input().split())

tudien = {}
for i in kv:
  x,y = i.split(":")
  if x in tudien:
    tudien[x].append(y)
  else:
    tudien[x] = [y]

print(tudien)


```

```python
#Bai 5
n = list(map(int,input().split()))
dic = {}
c1 = 0
c2 = 0
c3 = 0
for i in n:
  if i> 0: c1+=1
  elif i < 0: c2+=1
  else: c3+=1
dic["positives"] = c1
dic["negatives"] = c2
dic["zeros"] = c3
print(dic)

```

```python
#Bai 7
lst = list(map(int,input().split()))
print(lst[0])
print(lst[-1])
tup = []
for i in range(len(lst),0,-1):
  tup.append(str(lst[i-1]))
print(tuple(tup))
```

```python
#Bai 8
s = list(input().split())
dic = {}
for i in s:
  if i not in dic: dic[i] = 1
  else: dic[i]+=1
print(dic)


```

```python
#Bai 9
a= (input().strip("{}")).split(", ")
b= (input().strip("{}")).split(", ")
dic = {}
for i in a:
  x,y = i.split(": ")
  x = x.strip("'")
  if x not in dic: dic[x] = int(y)
for j in b:
  z,t = j.split(": ")
  z = z.strip("'")
  if z in dic: dic[z] += int(t)
  else: dic[z] = int(t)
print(dic)
```

```python
#Bai 10
lst = list(map(int,input().split()))
k = int(input())
finlst = []
for i in range(len(lst)):
  for j in range(i+1, len(lst)):
    if lst[i] + lst[j] == k:
      if tuple((lst[i],lst[j])) not in finlst:
        finlst.append(tuple((lst[i],lst[j])))
print(finlst)


```

```python
#Bai 11
lst = list(map(int,input().split()))
odd = []
even = []
for i in lst:
  if i % 2 == 0:
    even.append(i)
  else: odd.append(i)
print(tuple(even))
print(tuple(odd))
```

```python
#Bai 12
lst = list(map(int,input().split()))
maxlst=[]
rev1 = 0
for i in lst:
  if lst.count(i) > rev1:
    rev1 = lst.count(i)
    maxlst.clear()
    maxlst.append(i)
  elif lst.count(i) == rev1:
    maxlst.append(i)
print(min(maxlst))






```

```python
#Bai 13
lst = list(input().split())
dic = {}
for i in range(1,len(lst),2):
  for j in range(0,len(lst),2):
    if i -1  == j:
      dic[lst[i]] = lst[j]
print(dic)
```
