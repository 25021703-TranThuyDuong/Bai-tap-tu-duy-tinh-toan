```python
#W7A1
lst = list(map(int,input().split()))
n = int(input())
lst.sort()
def binary_search(n,lst):
  left, right= 0, len(lst) -1
  while left<=right:
    mid = (left + right)//2
    if lst[mid]==n:
      return mid
    elif lst[mid] < n: left = mid + 1
    else: right = mid - 1
  return -1
print(binary_search(n,lst))
```

```python
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    swapped = False
    for j in range(0,n-i+1):
      if arr[i] > arr[i-1]:
        arr[i], arr[i-1] = arr[i-1], arr[i]
        swapped = True
    if not swapped:
      break
def selection_sort(arr):
  for i in range(len(arr)):
    min_index = i
    for j in range(i+1,len(arr)):
      if arr[j] < min_index:
        min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]



```

```python
#W7A4
def fin_best(lst):
  sortedi = sorted(lst)
  dem = 0
  for i in sortedi:
    if dem < sortedi.count(i):
      dem = sortedi.count(i)
      best = i
  return best, dem
lst = list(map(int,input().split()))
a,b= fin_best(lst)
print(f"{a} xuat hien nhieu nhat, som nhat, {b} lan")
```

```python
#W7A8
def interval(lst,qu):
  n=[]
  lst.sort(key= lambda x : x[1]-x[0]+1)
  for q in qu:
    temp=-1
    for l,r in lst:
      if l <= q <= r:
        temp = r - l +1
        break
    n.append(temp)
  return n
lst = eval(input())
qu = eval(input())
print(interval(lst,qu))


```
