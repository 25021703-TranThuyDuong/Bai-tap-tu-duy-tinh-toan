```python
c#W8A1
class Cylinder:
  def __init__(self, bk, cc):
    self.bk = bk
    self.cc = cc
    self.pi = 3.14

  def dientichbemat(self):
    return 2*self.pi*self.bk*self.cc + 2*self.pi*self.bk*self.bk

  def thetich(self):
    return self.pi*self.bk*self.cc*self.cc
a,b = map(int,input().split())
p1 = Cylinder(a,b)
print(p1.dientichbemat())
print(p1.thetich())


```

```python
#W8A2
class Date:
  def __init__(self, day, month, year):
    self.day = day
    self.month = month
    self.year = year

  def is_leap_year(self):
    if self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0):
      return True
    else:
      return False

  def is_valid(self):
    if self.month < 1 or self.month > 12:
      return False
    days_in_month = [31, 29 if self.is_leap_year() else 28,\
                     31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if self.day < 1 or self.day > days_in_month[self.month - 1]:
      return False
    return True

  def next_date(self):
    days_in_month = [31, 29 if self.is_leap_year() else 28,\
                     31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    self.day += 1
    if self.day > days_in_month[self.month - 1]:
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

  def format(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"


try:
    d, m, y = map(int, input().split('/'))
    date = Date(d, m, y)

    if not date.is_valid():
        print("INVALID")
    else:
        date.next_date()
        print(date.format())

except:
    print("INVALID")
```

```python
#W8A4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def product(self):
        return self.a * self.b

    def divide(self):
        if self.b == 0:
            return "Không thể chia cho 0"
        return self.a / self.b

    def power(self):
        return self.a ** self.b

    def mod(self):
        if self.b == 0:
            return "Không thể chia cho 0"
        return self.a % self.b


# ===== MAIN =====
a,b = map(float,input().split())
calc = Calculator(a, b)

while True:
    print("\n===== MENU =====")
    print("1. Cộng")
    print("2. Trừ")
    print("3. Nhân")
    print("4. Chia")
    print("5. Lũy thừa")
    print("6. Chia dư")
    print("7. Nhập 2 số mới")

    choice = input("Chọn phép toán (1-7): ")

    if choice == "1":
        print("Kết quả:", calc.add())
    elif choice == "2":
        print("Kết quả:", calc.subtract())
    elif choice == "3":
        print("Kết quả:", calc.product())
    elif choice == "4":
        print("Kết quả:", calc.divide())
    elif choice == "5":
        print("Kết quả:", calc.power())
    elif choice == "6":
        print("Kết quả:", calc.mod())
    elif choice == "7":
        a = float(input("Nhập số thứ nhất mới: "))
        b = float(input("Nhập số thứ hai mới: "))
        calc.set_numbers(a, b)
    else:
        print("Lựa chọn không hợp lệ!")

    stop = input("Bạn có muốn thoát? ")
    if stop in ["yes", "YES", "Yes"]:
      break
    else:
      a,b = map(float,input('nhap lai 2 so: ').split())
      calc.set_numbers(a, b)


```
