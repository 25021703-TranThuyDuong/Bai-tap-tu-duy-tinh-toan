```python
#Bai 1
def Max_Of_Two(a,b):
  max = a
  if b > max:
    max = b
  return max
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#Bai 2
def swap(a,b):
  a,b=b,a
  return a,b
a,b=map(int,input().split())
a,b=swap(a,b)
print(a,b)
```

```python
#Bai 3
def prime(x):
  for i in range(2,int(x**0.5)+1):
    if x%i==0:
      return False
  return x >= 2
print(prime(int(input())))
```

```python
#Bai 4
def pernum(x):
  tong=0
  for i in range(1,x):
    if x%i==0:
      tong+=i
  if tong == x: return True
  return False
print(pernum(int(input())))
```

```python
#Bai 5
def location(arr,k):
  for i in range(1,len(arr)):
    if arr[i] == k:
      return i+1
  return -1
arr = list(map(int,input().split()))
k = int(input())
print(location(arr,k))
```

```python
#Bai 6
def fac(n):
  gt=1
  for i in range(1,n+1):
    gt*=i
  return gt
print(fac(int(input())))
```

```python
#Bai 7
def cal(a,b,operat):
  if operat == '+': return a+b
  elif operat == '-': return a-b
  elif operat == '*': return a*b
  elif operat == '/': return a/b
a,operat,b=input().split()
a,b=int(a),int(b)
print(f"{a} {operat} {b} = {cal(a,b,operat)}")
```
