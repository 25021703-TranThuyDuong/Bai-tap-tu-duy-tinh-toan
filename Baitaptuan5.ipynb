```python
#Bai 1
def Max_Of_Two(a,b):
  max = a
  if b > max:
    max = b
  return max
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#Bai 2
def swap(a,b):
  a,b=b,a
  return a,b
a,b=map(int,input().split())
a,b=swap(a,b)
print(a,b)
```

```python
#Bai 3
def prime(x):
  for i in range(2,int(x**0.5)+1):
    if x%i==0:
      return False
  return x >= 2
print(prime(int(input())))
```

```python
#Bai 4
def pernum(x):
  tong=0
  for i in range(1,x):
    if x%i==0:
      tong+=i
  if tong == x: return True
  return False
print(pernum(int(input())))
```

```python
#Bai 5
def location(arr,k):
  for i in range(1,len(arr)):
    if arr[i] == k:
      return i+1
  return -1
arr = list(map(int,input().split()))
k = int(input())
print(location(arr,k))
```

```python
#Bai 6
def fac(n):
  gt=1
  for i in range(1,n+1):
    gt*=i
  return gt
print(fac(int(input())))
```

```python
#Bai 7
def cal(a,b,operat):
  if operat == '+': return a+b
  elif operat == '-': return a-b
  elif operat == '*': return a*b
  elif operat == '/': return a/b
a,operat,b=input().split()
a,b=int(a),int(b)
print(f"{a} {operat} {b} = {cal(a,b,operat)}")
```

```python
 #Bai 8
def hamming(a,b):
  dis = 0
  s1 = bin(a)[2:]
  s2 = bin(b)[2:]
  s1 = s1[::-1]
  s2 = s2[::-1]

  while len(s1) > len(s2):
      s2 += "0"
  while len(s1) < len(s2):
      s1 += "0"
  for i in range(len(s1)):
    if s1[i] != s2[i]:
      dis+=1
  return dis, s1, s2
a,b= map(int,input().split())
print(hamming(a,b))
```

```python
#Bai 9
def tongchuso(x):
  tong=0
  while x > 0:
    tong+= x%10
    x//=10
  return tong
print(tongchuso(int(input())))
```

```python
def anhxa(a,b):
  map1 = {}
  map2 = {}
  if len(a) != len(b): return False
  for i in range(len(a)):
    c1,c2 = a[i], b[i]
    if c1 in map1:
      if map1[c1] != c2: return False
    else: map1[c1] = c2
    if c2 in map2:
      if map2[c2] != c1: return False
    else: map2[c2] = c1
  return True
a = input()
b = input()
print(anhxa(a,b))
```
