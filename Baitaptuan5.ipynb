```python
#Bai 1
def Max_Of_Two(a,b):
  max = a
  if b > max:
    max = b
  return max
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

```python
#Bai 2
def swap(a,b):
  a,b=b,a
  return a,b
a,b=map(int,input().split())
a,b=swap(a,b)
print(a,b)
```

```python
#Bai 3
def prime(x):
  for i in range(2,int(x**0.5)+1):
    if x%i==0:
      return False
  return x >= 2
print(prime(int(input())))
```
