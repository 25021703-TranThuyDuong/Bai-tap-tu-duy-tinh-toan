```python
#W11A3
def tongbinhphuong(n):
  tong=0
  while n!=0:
    tong+=(n%10)**2
    n//=10
  return tong
try:
  a = int(input())
  check=set()
  while a!=1 and a not in check:
    check.add(a)
    a=tongbinhphuong(a)
  if a==1:
    print('YES')
  else:
    print('NO')


except ValueError:
  print('NO')



```
